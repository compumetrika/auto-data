```python
from fredapi import Fred
import pylab as plt
fred = Fred(api_key='')
print("This is my API key, please register your own here: https://research.stlouisfed.org/docs/api/api_key.html")
import numpy as np
import pandas as pd
```

```python
data = fred.get_series('SP500')
gdp  = fred.get_series('GDP')
```

```python
plt.plot(data['2007-01-1':])
#plt.plot(gdp['2007-01-1':])
plt.show()
```

```python
plt.plot(gdp['2007-01-1':])
plt.show()
```

```python
plt.plot(np.diff(np.log(gdp['2007-01-1':]))*4*100)
plt.plot(np.repeat(np.mean(np.diff(np.log(gdp['2007-01-1':]))*4*100), len(np.diff(np.log(gdp['2007-01-1':]))*4)), 'k--')
plt.show()
```

```python
fig, ax1 = plt.subplots()
ax1.plot(gdp['2007-01-1':], linestyle='solid', color='black')

ax2 = ax1.twinx()
ax2.plot(data['2007-01-1':], linestyle='dashed')

fig.tight_layout()
plt.show()
```
