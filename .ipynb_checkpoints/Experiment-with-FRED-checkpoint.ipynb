```python
from fredapi import Fred
import pylab as plt
import numpy as np
import pandas as pd

with open("./to_exclude/key","r") as f:
    my_key = f.readline()
    my_key.rstrip()

fred = Fred(api_key=my_key)
print("NOTE: You must obtain your own API key from FRED. Please register here to do so: https://research.stlouisfed.org/docs/api/api_key.html")
```

```python
data = fred.get_series('SP500')
gdp  = fred.get_series('GDP')
```

```python
start_date = '2008-01-1'
plt.plot(data[start_date:])
plt.show()
```

```python
plt.plot(gdp[start_date:])
plt.show()
```

```python
plt.plot(np.diff(np.log(gdp[start_date:]))*4*100)
plt.plot(np.repeat(np.mean(np.diff(np.log(gdp[start_date:]))*4*100), len(np.diff(np.log(gdp[start_date:]))*4)), 'k--')
plt.show()
```

```python
fig, ax1 = plt.subplots()
ax1.plot(gdp[start_date:], linestyle='solid', color='black')

ax2 = ax1.twinx()
ax2.plot(data[start_date:], linestyle='dashed')

fig.tight_layout()
plt.show()
```
